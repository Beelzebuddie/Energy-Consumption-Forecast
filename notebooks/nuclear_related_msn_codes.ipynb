```python
import pandas as pd
```

```python
# Load the provided Excel file
file_path = r"C:\Code\energy-demand-forecasting\docs\msn_codes_and_descriptions.xlsx"

# Skip additional rows to locate the start of the MSN data in the "MSN descriptions" sheet
msn_descriptions = pd.read_excel(file_path, sheet_name="MSN descriptions", header=None, skiprows=20)

# Display the first few rows to check the structure
msn_descriptions.head(10)
```

```python
# Rename the columns for clarity
msn_descriptions.columns = ["Index", "MSN", "Description", "Unit"]

# Drop unnecessary rows (e.g., empty rows)
msn_descriptions = msn_descriptions.dropna(subset=["MSN", "Description"])

# Filter MSN codes for nuclear-related terms
keywords = ["nuclear", "electric", "uranium", "generation"]
nuclear_msn = msn_descriptions[msn_descriptions['Description'].str.contains('|'.join(keywords), case=False, na=False)]

# Display nuclear-related MSN codes and descriptions
nuclear_msn.reset_index(drop=True, inplace=True)
nuclear_msn
```

```python
# Save the file to the specified path
specified_path = r"C:\Code\energy-demand-forecasting\docs\nuclear_related_msn_codes.csv"
nuclear_msn.to_csv(specified_path, index=False)

specified_path

```
